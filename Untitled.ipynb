```python
import numpy as np
from matplotlib import pyplot as plt
import math

EPS_START = 0.9
EPS_END = 0.05
EPS_DECAY = 100

steps_list = []
eps_list = []

for steps_done in range (1000):
    eps_threshold = EPS_END + (EPS_START - EPS_END) * math.exp(-1. * steps_done / EPS_DECAY)
    eps_list.append(eps_threshold)
    steps_list.append(steps_done)

plt.plot(steps_list,eps_list)
    
```
